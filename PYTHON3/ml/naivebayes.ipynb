```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("C:/Users/Manoj/Documents/ml_record/tennis.csv")
data
```

```python
print(data['outlook'],end='\n-----------------------')
print(data['temp'],end='\n-----------------------')
print(data['humidity'],end='\n-----------------------')
print(data['wind'],end='\n-----------------------')
print(data['play'],end='\n-----------------------')
```

```python
outlook = data.groupby(['outlook', 'play']).size()
temp = data.groupby(['temp', 'play']).size()
humidity = data.groupby(['humidity', 'play']).size()
wind = data.groupby(['wind', 'play']).size()
play = data.play.value_counts()
```

```python
print(outlook)
```

```python
total_prob=play['Yes']+play['No']
def find_probablity(oval,tval,hval,wval,rval):
    out_val=outlook[oval][rval]/play[rval]
    temp_val=temp[tval][rval]/play[rval]
    wind_val=wind[wval][rval]/play[rval]
    humidity_val=humidity[hval][rval]/play[rval]
    play_val=play[rval]/total_prob
    return out_val*temp_val*wind_val*humidity_val*play_val

    
def predict_play(oval,tval,hval,wval):
    p_no=0
    p_yes=find_probablity(oval,tval,hval,wval,'Yes')
    if oval!='Overcast':
        p_no=find_probablity(oval,tval,hval,wval,'No')
    print(p_yes,p_no)
    print('You Can Play ') if p_yes>p_no else print('You cannot play')
```

```python
predict_play('Overcast','Hot','High','Weak')
```
